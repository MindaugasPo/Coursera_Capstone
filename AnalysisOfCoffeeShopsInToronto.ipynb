# Analysis of coffee shops in Toronto

## Introduction

This report describes a study of Toronto coffe shops. This topic is important for individuals or companies that manage, trying to optimize or are considering opening coffee shops in Toronto area. The study will try to answer question what are best locations to own coffee shops in Toronto to avoid high competition.

## Business case

If someone wants to open a coffee, natural question arrises - where to do it in order to optimize profitability. In order to answer that, one should take into account many variable like nearby competition, amount of people living or visiting the area, type of neighborhood (business district, living house district, schools, universities, etc.), average income of people that live or visit the area and many more.
This report limits analysis to one variable - the amount of competition in surrounding areas. It will attempt to answer question - what are areas with least number of coffe shops.
